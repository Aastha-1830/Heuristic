```python
# define list of different colours 
colors=["red", "gold", "cyan", "blue", "magenta", "orange", "yellow", "green", "white", "brown", "violet", "sienna", "purple", "wheat1", "salmon", "pink", "darkorange", "silver", "coral", "maroon", "olive", "royalblue" ]
print(len(colors))
```

```python
# define sequence in which you select route
routes = ['M1', 'M2', 'M4', 'M5', 'M6', 'M8', 'M10', '12', 'M13', '16', 'M17', '18', '21', '27', '37', '50', '60', '61', '62', '63', '67', '68']
```

```python
# Empty dictionary to store route and its colour
assign_dict = {'M1': '', 'M2': '', 'M4': '', 'M5': '', 'M6': '', 'M8': '', 'M10': '', '12': '', 'M13': '', '16': '', 'M17': '', '18': '', '21': '', '27': '', '37': '', '50': '', '60': '', '61': '', '62': '', '63': '', '67': '', '68': ''}
print(assign_dict)
```

```python
#code using the heuristic approach
N=len(df) 
length = len(routes)
print("Number of routes:", N)
l_color = 0
assign_dict['M1']=colors[l_color]
for i in range(length):
  # list of feasible routes
  feasible_colors = colors.copy()
  # write code here for checking for intersection
  for j in range(i):
    denote = 0
    if(df.iat[i,j+1]!=1):        #checking the intersection of i, j routes
      col = assign_dict[routes[j]]
      count=0
      for k in range(i):
        if col==assign_dict[routes[k]]:
          if df.iat[i,k+1]==1:  #if intersection, assign different color
            count+=1
      if count==0:
        denote = 1
        assign_dict[routes[i]] = col
        break
    if denote == 0 and j==i-1:
      l_color+=1
      assign_dict[routes[i]]=colors[l_color]

  # write code here for deleting some colours from feasible_colours
feasible_colors=feasible_colors[0:l_color+1]  

  # write code here for storing route and its colour in Assignment dictionary
print("List of colors assigned to routes:", assign_dict)  
print("Remaining colours:", feasible_colors)
print("Number of colours needed are:", len(feasible_colors))
```
